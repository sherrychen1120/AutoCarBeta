```python
import cv2
import numpy as np
import glob
import sys
import matplotlib.pyplot as plt
%matplotlib inline
import time
```

```python
print ("OpenCV:",  cv2.__version__)
print ("Numpy : ", np.__version__)
print ("Python:",  sys.version)
```

```python
# load training data
dim = 120*320
X = np.empty((0, dim))
y = np.empty((0, 4))
training_data = glob.glob('training_data/1536083382.npz')
single_npz = np.load(training_data[0]);
len(single_npz['train'][0])
```

```python
for single_npz in training_data:
    with np.load(single_npz) as data:
        train = data['train']
        train_labels = data['train_labels']
    X = np.vstack((X, train))
    y = np.vstack((y, train_labels))

print ('Image array shape: ', X.shape)
print ('Label array shape: ', y.shape)
```

```python
fig=plt.figure(figsize=(54, 30))
columns = 3
rows = 9
for i in range(1,7):
    fig.add_subplot(rows, columns, i)
    plt.imshow(X[i-1].reshape(120, 320), cmap='gray')
plt.show()
```

```python
# Script used for selecting part of the collected data

X_new = np.empty((0, 120 * 320))
y_new = np.empty((0, 4))
for i in range(1,7):
    X_new = np.vstack((X_new, X[i-1]))
    y_new = np.vstack((y_new, y[i-1]))

file_name = str(int(time.time()))
directory = "/Users/sherrychen/Documents/UPenn/EC/AutoRCCar Project/my_rccar/computer/training/training_data"
np.savez(directory + '/' + file_name + '.npz', train=X_new, train_labels=y_new)
```
